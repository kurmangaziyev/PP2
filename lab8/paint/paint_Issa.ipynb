# Приложение Paint на Pygame

Этот Jupyter Notebook содержит две версии приложения Paint, реализованные с использованием Pygame.

## Функциональность:
- Рисование линий
- Рисование прямоугольников
- Рисование кругов
- Ластик
- Выбор цвета

Вторая версия расширяет первую, добавляя больше фигур:
- Квадрат
- Равносторонний треугольник
- Прямоугольный треугольник
- Ромб


```
import pygame

pygame.init()

# Устанавливаем цвета
WHITE = (255,255,255)
RED = (255,0,0)
GREEN = (0,255,0)
BLUE = (0,0,255)
BLACK = (0,0,0)

COLOR = RED

# Настройки FPS
clock = pygame.time.Clock()
FPS = 30

# Переменные
prev, cur = None, None

# Настройки экрана
length, width = 1000, 600
screen = pygame.display.set_mode((length, width))
screen.fill(WHITE)
running = True

pen = "mouse"

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # Выбор режима рисования
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_q:
                pen = "mouse"
            if event.key == pygame.K_w:
                pen = "rect"
            if event.key == pygame.K_e:
                pen = "circle"
            if event.key == pygame.K_r:
                pen = "eraser"

            # Выбор цвета
            if event.key == pygame.K_z:
                COLOR = RED
            if event.key == pygame.K_x:
                COLOR = GREEN
            if event.key == pygame.K_c:
                COLOR  = BLUE
            if event.key == pygame.K_v:
                COLOR = BLACK

        # Рисование карандашом
        if pen == "mouse":
            if event.type == pygame.MOUSEBUTTONDOWN:
                prev = pygame.mouse.get_pos()
            if event.type == pygame.MOUSEMOTION and prev:
                cur = pygame.mouse.get_pos()
                pygame.draw.line(screen, COLOR, prev, cur, 1)
                prev = cur
            if event.type == pygame.MOUSEBUTTONUP:
                prev = None

    pygame.display.flip()
    clock.tick(FPS)

pygame.quit()
```
