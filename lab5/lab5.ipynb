```python

def find_string(text):
    matches = re.sub(r'[' ',.]', ':', text)
    return matches


with open("row.txt", "r") as file:
    text = file.read()

result = find_string(text)
print("String found:", result)
```

```python

def camel(text):
    matches = re.sub(r'_([a-zа-я])', lambda match: match.group(1).upper(), text)
    return matches


with open("row.txt", "r") as file:
    text = file.read()

result = camel(text)
print("String found:", result)
```

```python
import re

def split_at_uppercase(text):
    return re.split(r'(?=[A-Z])', text)

# Read from file
with open("row.txt", "r", encoding="utf-8") as file:
    text = file.read()

result = split_at_uppercase(text)
print("Split result:", result)
```

```python

import re

def split_at_uppercase(text):
    return re.sub(r'(?<!^)(?=[A-Z])', ' ', text)

# Read from file
with open("row.txt", "r") as file:
    text = file.read()

result = split_at_uppercase(text)
print("Split result:", result)
```

```python

def snake(text):
    res = re.sub(r'(?<=\w)(?=[А-ЯA-Z])', '_', text)
    return res.lower().strip('_').replace('__', '_')


with open("row.txt", "r") as file:
    text = file.read()

result = snake(text)
print("String found:", result)
```
