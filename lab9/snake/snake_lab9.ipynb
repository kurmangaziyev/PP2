
# Игра "Змейка"

## Описание
"Змейка" — это классическая игра, в которой игрок управляет змейкой, движущейся по экрану. 
Цель игры — съесть как можно больше еды, избегая столкновений со стенами и собственным хвостом.

## Функциональность Lab 8:
- Проверка на столкновение со стенами и выход за границы экрана
- Генерация еды в случайной позиции (не на змее и не на стенах)
- Добавление уровней (увеличение скорости после 3-4 съеденных объектов)
- Отображение текущего счета и уровня

## Дополнения в Lab 9:
- Разные типы еды с разным "весом" (1, 2, 3 очка за каждую)
- Еда исчезает через определенное время, если не съедена


```

import pygame
import random

# Инициализация Pygame
pygame.init()

# Настройки экрана
screen_width = 600
screen_height = 400
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Snake Game - Lab 9")

# Цвета
white = (255, 255, 255)
black = (0, 0, 0)
red = (255, 0, 0)
green = (0, 255, 0)
blue = (0, 0, 255)

# Настройки змеи
block_size = 20
snake_speed = 10

# Шрифты
font = pygame.font.SysFont("Verdana", 20)

# Функция для отрисовки змеи
def draw_snake(snake_list):
    for block in snake_list:
        pygame.draw.rect(screen, green, [block[0], block[1], block_size, block_size])

# Функция генерации еды
def generate_food(snake_list):
    while True:
        food_x = round(random.randrange(0, screen_width - block_size) / 20.0) * 20.0
        food_y = round(random.randrange(0, screen_height - block_size) / 20.0) * 20.0
        food_weight = random.choice([1, 2, 3])  # Разные веса еды (1, 2, 3 очка)
        if [food_x, food_y] not in snake_list:
            return food_x, food_y, food_weight

# Основной игровой цикл
def game_loop(snake_speed):
    game_over = False
    game_close = False

    snake_list = []
    snake_length = 1
    snake_x = screen_width / 2
    snake_y = screen_height / 2
    snake_x_change = 0
    snake_y_change = 0

    # Генерация первой еды
    food_x, food_y, food_weight = generate_food(snake_list)
    food_timer = 300  # Таймер для еды (примерно 5 секунд)

    score = 0
    level = 1

    clock = pygame.time.Clock()

    while not game_over:
        while game_close:
            screen.fill(white)
            message = font.render("You lost! Press Q to quit or C to play again", True, red)
            screen.blit(message, [screen_width / 6, screen_height / 3])
            pygame.display.update()

            for event in pygame.event.get():
                if event.type == pygame.QUIT:
                    game_over = True
                    game_close = False
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = True
                        game_close = False
                    if event.key == pygame.K_c:
                        game_loop(snake_speed)

        # Обработка событий управления
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT and snake_x_change == 0:
                    snake_x_change = -block_size
                    snake_y_change = 0
                elif event.key == pygame.K_RIGHT and snake_x_change == 0:
                    snake_x_change = block_size
                    snake_y_change = 0
                elif event.key == pygame.K_UP and snake_y_change == 0:
                    snake_y_change = -block_size
                    snake_x_change = 0
                elif event.key == pygame.K_DOWN and snake_y_change == 0:
                    snake_y_change = block_size
                    snake_x_change = 0

        snake_x += snake_x_change
        snake_y += snake_y_change

        # Проверка на выход за границы экрана
        if snake_x >= screen_width or snake_x < 0 or snake_y >= screen_height or snake_y < 0:
            game_close = True

        # Проверка на столкновение с едой
        if snake_x == food_x and snake_y == food_y:
            score += food_weight  # Увеличиваем счет на значение веса еды
            snake_length += food_weight  # Увеличиваем длину змеи
            food_x, food_y, food_weight = generate_food(snake_list)
            food_timer = 300  # Сбрасываем таймер еды

            if score % 5 == 0:  # Повышение уровня каждые 5 очков
                level += 1
                snake_speed += 2

        # Таймер еды (уменьшаем каждую итерацию)
        food_timer -= 1
        if food_timer <= 0:
            food_x, food_y, food_weight = generate_food(snake_list)
            food_timer = 300

        screen.fill(white)
        pygame.draw.rect(screen, red if food_weight == 1 else blue, [food_x, food_y, block_size, block_size])

        snake_head = [snake_x, snake_y]
        snake_list.append(snake_head)

        if len(snake_list) > snake_length:
            del snake_list[0]

        # Проверка на столкновение змеи с самой собой
        for block in snake_list[:-1]:
            if block == snake_head:
                game_close = True

        draw_snake(snake_list)

        # Отображение счета и уровня
        score_text = font.render(f"Score: {score}", True, black)
        level_text = font.render(f"Level: {level}", True, black)
        screen.blit(score_text, (10, 10))
        screen.blit(level_text, (10, 30))

        pygame.display.update()
        clock.tick(snake_speed)

    pygame.quit()

# Запуск игры
game_loop(snake_speed)
```
