# Объединенный код для игры

```
import pygame
import random

# zdes funkciji dlya igri kotorie otvechayut za moneti igroka i vraga

def create_coin(length, width, coin_radius):
    return pygame.Rect(random.randint(0, length - coin_radius * 2),
                       random.randint(0, width - coin_radius * 2),
                       coin_radius * 2, coin_radius * 2)

def draw_player(screen, player_pos):
    pygame.draw.rect(screen, (0, 0, 255), player_pos)  # Рисуем игрока как синий прямоугольник

def draw_coins(screen, coins):
    for coin in coins:
        pygame.draw.ellipse(screen, (255, 255, 0), coin)  # Желтые круги

def collect_coins(player_pos, coins):
    collected = 0
    for coin in coins[:]:  # Перебираем копию списка
        if player_pos.colliderect(coin):  # Проверяем столкновение
            coins.remove(coin)  # Убираем монету
            collected += 1
    return collected

def draw_coin_count(screen, collected_coins, length):
    font = pygame.font.SysFont(None, 36)
    text = font.render(f"Coins: {collected_coins}", True, (0, 0, 0))
    screen.blit(text, (length - 150, 10))

def move_enemy(enemy_pos, enemy_speed, length):
    enemy_pos.x += enemy_speed
    if enemy_pos.right >= length or enemy_pos.left <= 0:
        enemy_speed = -enemy_speed  # Изменяем направление движения
    return enemy_pos, enemy_speed
```

```
import pygame
import random
from components import create_coin, draw_player, draw_coins, collect_coins, draw_coin_count, move_enemy

pygame.init()

WHITE = (255, 255, 255)
RED = (255, 0, 0)
BLUE = (0, 0, 255)

clock = pygame.time.Clock()
FPS = 30

length, width = 1000, 600
player_pos = pygame.Rect(400, 300, 50, 50)
player_speed = 5
coins = []
collected_coins = 0
coin_radius = 10
enemy_speed = 2
enemy_pos = pygame.Rect(random.randint(0, length-50), random.randint(0, width-50), 50, 50)

screen = pygame.display.set_mode((length, width))
screen.fill(WHITE)
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_pos.left > 0:
        player_pos.x -= player_speed
    if keys[pygame.K_RIGHT] and player_pos.right < length:
        player_pos.x += player_speed
    if keys[pygame.K_UP] and player_pos.top > 0:
        player_pos.y -= player_speed
    if keys[pygame.K_DOWN] and player_pos.bottom < width:
        player_pos.y += player_speed

    if random.random() < 0.05:
        coin = create_coin(length, width, coin_radius)
        coins.append(coin)

    collected_coins += collect_coins(player_pos, coins)

    if collected_coins >= 5 and collected_coins % 5 == 0:
        enemy_speed += 1

    enemy_pos, enemy_speed = move_enemy(enemy_pos, enemy_speed, length)

    screen.fill(WHITE)
    draw_player(screen, player_pos)
    draw_coins(screen, coins)
    draw_coin_count(screen, collected_coins, length)
    pygame.draw.rect(screen, RED, enemy_pos)
    pygame.display.flip()

    clock.tick(FPS)

pygame.quit()
```
